# Определить риск для компании на основе аудита

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data_audit = pd.read_csv('audit_data/audit_risk.csv')
```

```python
data_audit.head()
```

```python
data_audit.info()
```

```python
data_audit.corr()
```

```python
X = data_audit.dropna().drop(['Risk','LOCATION_ID'], axis=1)
Y = data_audit.dropna().Risk
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=123)
```

```python
clf = SVC(gamma='auto')
clf.fit(x_train, y_train)
```

```python
pred = clf.predict(x_test)
```

```python
print("The accuracy is: ", accuracy_score(y_test, pred))
```
